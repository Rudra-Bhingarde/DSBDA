```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('forestfires.csv')
data
```

```python
plt.figure()
plt.hist(data['area'], bins=30)
plt.xlabel('Burned Area')
plt.ylabel('Frequency')
plt.title('Distribution of Burned Area')
plt.show()
```

```python
plt.figure()
plt.scatter(data['temp'], data['area'])
plt.xlabel('Temperature')
plt.ylabel('Burned Area')
plt.title('Temperature vs Burned Area')
plt.show()
```

```python
plt.figure()
plt.scatter(data['wind'], data['area'])
plt.xlabel('Wind Speed')
plt.ylabel('Burned Area')
plt.title('Wind Speed vs Burned Area')
plt.show()
```
